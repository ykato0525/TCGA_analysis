```python
def find_txt_file(directory_path):
    for file in os.listdir(directory_path):
        if file.endswith(".txt"):
            return os.path.join(directory_path, file)
    return None
```

```python
directory_path = "./" + str(case_list[0])
print(directory_path)
```

```python
txt_f = find_txt_file(directory_path)
```

```python
for i in range(1,len(case_list)):
    directory_path = "./" + str(case_list[i])
    txt_f = find_txt_file(directory_path)
    print(txt_f)
    if txt_f != None:
        df = pd.read_table(txt_f, sep="\t", header=None, index_col=0)
        join_df = df.loc[cgas_promoters]
        join_df.rename(columns={1:case_list[i]}, inplace=True)
        filtered_df = filtered_df.join(join_df)
filtered_df.to_csv("TCGA_BRCA_CGAS_MET.csv")           
```
